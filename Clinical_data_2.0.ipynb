```python
data.gender.info()

#things to correct different formats
#no missing entries
data.gender.unique()
```

```python
data.gender = data.gender.replace(['m',"male", "Male"], 'M')
data.gender = data.gender.replace(['f',"female", "Female"], 'F')
data.gender.unique()
```

```python
data.name
#so the issue is the names are unevenly in samll and capital letters , and some are missing
# firstly dropping the columns, where the patient name is missing
```

```python
data.dropna(subset = ["name"], inplace = True)
```

```python
data.info()
#nan columns of the "name" has been dropped
```

```python
# organising the names in the correct format
data["name"] = data["name"].str.title()
data.name
```

```python




#data.age.info()#so age is an object
#data.astype({"age":"int"})
data["age"]= pd.to_numeric(data["age"], errors = "coerce")
```

```python
data.age.info()
```

```python
import numpy as np

data.loc[(data["age"]>200)|(data["age"]<1), "age"] = np.nan
```

```python
data["age"] =  data["age"].fillna(data["age"].median())
```

```python
data.age.unique()
#no impossible ages, considering valid ages upto """200"""
```

```python
# next tasks
# handle missing blood group bill , and satisfaction score
# convert visit date and follow up date to date time
# check and remove duplicate
# final df info to check everything is all right \
```

```python
# some blood groups are missing so we will replace them to not available values
data['blood_group'] = data["blood_group"].fillna("not_available")
data['satisfaction_score'] = data["satisfaction_score"].fillna("not_available")
```

```python
data['satisfaction_score'].unique()
```

```python
data["blood_group"].unique()
```

```python
#converting visit and follow up dsate to date and time
data["visit_date"] = pd.to_datetime(data["visit_date"])
data["follow_up_date"] = pd.to_datetime(data["follow_up_date"])
```

```python
data.info()
#date and time have been changed into date and time format
```

```python
""" to check and remove duplicates """
data.loc[data.duplicated(keep = False)]#no duplicates
```

```python
# now analysing the cleaned data
```

```python
#which doctor has the highest satisfaction score
#score = data.groupby("doctor")["satisfaction_score"].mean()

#satisfaction score need to be converted into int
data["satisfaction_score"]= pd.to_numeric(data["satisfaction_score"], errors='coerce' )
data.satisfaction_score.info()
```

```python
score = data.groupby("doctor")["satisfaction_score"].mean().sort_values(ascending = False)

score
#so Dr.Iyer has maximum satisfaction score
```

```python
#avg consultation bill per department
bill = data.groupby("department")["consultation_bill"].mean().sort_values(ascending = False)
bill
```

```python
#scatter plot; BMI V\S blood sugar
bmi = data["bmi"]
blood_sugar= data['blood_sugar']
plt.scatter(bmi, blood_sugar, color ="green", s = 5)
plt.xlabel('Patients BMI (kg/m^2)')

plt.ylabel("Blood Sugar (mg/dl)")
plt.title("BMI v/s Blood Sugar")
plt.tight_layout()
plt.savefig('catter.png', dpi = 300)
```

```python
#creating new column by bp categhory
def bp_category(row):
  if row['bp_systolic']<= 120 :
    return"Normal"
  elif row["bp_systolic"] >= 190:
    return"High"
  else:
    return"Elevated"

data["bp_category"] = data.apply(bp_category, axis = 1)
data.info()
```
